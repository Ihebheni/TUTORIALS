# <font color=red> Importation des outils et des données

Données : Prix de clôture de l'action de Google ('GOOG') de 2005 à 2020.

```python
# Outils

import pandas as pd

from pandas_datareader.data import DataReader

import matplotlib.pyplot as plt
```

OHLC (Open High Low Close)

```python
# Importation des données

df = DataReader('GOOG', 'yahoo', start = '2005')

# 5 premières lignes

df.head()
```

# <font color=red> Moyennes mobiles

```python
# Moyennes mobiles sur 50 jours

df['MA 50'] = df['Close'].rolling(50).mean()

# Moyennes mobiles sur 100 jours

df['MA 100'] = df['Close'].rolling(100).mean()
```

# <font color=red> Visualisation des séries

```python
# Variables qu'on veut représenter

vars = ['Close', 'MA 50', 'MA 100']

# Graphique

df[vars].plot(figsize = (8, 5));
```
