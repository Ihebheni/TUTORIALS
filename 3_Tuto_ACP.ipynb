# ***Librairies***

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

# ***Importation des données***

```
# Noms des colonnes dans wine.data
wines_cols = ['class_label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 
              'Total phenols', 'Flavanoids', 'Nonflavanoid phenols','Proanthocyanins', 
              'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']

df = pd.read_csv('/content/drive/My Drive/TUTORIELS/ACP/wine.data', header=None, names=wines_cols)
df
```

```
df.info()
```

```
df.drop(columns=['class_label'], inplace=True)
df
```

```
df_array = df.values

pipe = Pipeline([('scaler', StandardScaler()),
                 ('acp', PCA(n_components=0.8))])
```

```
df_pca = pipe.fit_transform(df_array)
df_pca
```

```
df_pca.shape
```

```
df_pca_data = pd.DataFrame(df_pca, columns=['PC1','PC2', 'PC3', 'PC4', 'PC5'])
df_pca_data
```

```
df_all = pd.concat([df, df_pca_data], axis=1)
df_all
```
